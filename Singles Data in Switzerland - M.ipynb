```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import dateutil.parser
```

# Feeding in the data for ledige Frauen

```python
#Total and 20-39
df_M_39 = pd.read_csv('data/2015_M_TOTAL_20bis39.csv', encoding='latin', sep='\t', skiprows=1)
#40-64
df_M_64 = pd.read_csv('data/2015_M_TOTAL_40bis64.csv', encoding='latin', sep='\t', skiprows=1)
#65-89
df_M_89 = pd.read_csv('data/2015_M_TOTAL_65bis89.csv', encoding='latin', sep='\t', skiprows=1)
#90-100+
df_M_100 = pd.read_csv('data/2015_M_TOTAL_90bis100+.csv', encoding='latin', sep='\t', skiprows=1)
```

```python
#Deleting the columns I don' need
del df_M_64['Kanton (-) / Bezirk (>>) / Gemeinde (......)']
del df_M_64['Bevölkerungstyp']
del df_M_64['Geschlecht']
del df_M_64['Zivilstand']
del df_M_64['Jahr']
```

```python
del df_M_89['Kanton (-) / Bezirk (>>) / Gemeinde (......)']
del df_M_89['Bevölkerungstyp']
del df_M_89['Geschlecht']
del df_M_89['Zivilstand']
del df_M_89['Jahr']
```

```python
df_M_100.head()
del df_M_100['Kanton (-) / Bezirk (>>) / Gemeinde (......)']
del df_M_100['Bevölkerungstyp']
del df_M_100['Geschlecht']
del df_M_100['Zivilstand']
del df_M_100['Jahr']
```

```python
df = pd.concat([df_M_39, df_M_64, df_M_89, df_M_100], axis=1, join='inner')
```

```python
df
```

```python
#Creating DF for total
df_total_2015 = df[['Kanton (-) / Bezirk (>>) / Gemeinde (......)', 'Geschlecht', 'Altersklasse - Total']].copy()
#Filerting out every fourth row, starting with the first
df_total_2015 = df_total_2015.iloc[::4, :]
#Renaming Columns
df_total_2015.columns = [['Gemeinde', 'Geschlecht', 'Männer - Total']]
```

```python
#Creating DF for ledig
df_alle_2015 = df[['Kanton (-) / Bezirk (>>) / Gemeinde (......)', 'Geschlecht', 'Zivilstand', 'Altersklasse - Total']].copy()
```

```python
#Slicing all of the relevant rows
df_ledig_2015 = df_alle_2015.iloc[1::4, :]
df_verwitwet_2015 = df_alle_2015.iloc[2::4, :]
df_geschieden_2015 = df_alle_2015.iloc[3::4, :]
```

```python
#Merging for full count of singles incl. verwitwet & geschieden
df_ledig_2015 = df_ledig_2015.merge(df_verwitwet_2015, left_on='Kanton (-) / Bezirk (>>) / Gemeinde (......)', right_on='Kanton (-) / Bezirk (>>) / Gemeinde (......)')
df_ledig_2015 = df_ledig_2015.merge(df_geschieden_2015, left_on='Kanton (-) / Bezirk (>>) / Gemeinde (......)', right_on='Kanton (-) / Bezirk (>>) / Gemeinde (......)')
```

```python
del df_ledig_2015['Geschlecht_x']
del df_ledig_2015['Zivilstand_x']
del df_ledig_2015['Geschlecht_y']
del df_ledig_2015['Zivilstand_y']
del df_ledig_2015['Geschlecht']
del df_ledig_2015['Zivilstand']
```

```python
df_ledig_2015.columns = [['Gemeinde', 'Ledig', 'Verwitwet', 'Geschieden']]
```

```python
df_ledig_2015['Total ledige Männer ab 20'] = df_ledig_2015.sum(axis=1)
```

```python
df_ledig_2015.head()
```

```python
df_total_2015.head()
```

```python
df_total_2015 = df_total_2015.merge(df_ledig_2015, left_on='Gemeinde', right_on='Gemeinde')
```

```python
del df_total_2015['Ledig']
del df_total_2015['Verwitwet']
del df_total_2015['Geschieden']
```

```python
df_total_2015['Anteil ledige Männer'] = round(df_total_2015['Total ledige Männer ab 20'] / df_total_2015['Männer - Total'] * 100, 1)
```

# Now the age categories

```python
#Slicing
df_all = df.iloc[::4, :]
df_ledig_all = df.iloc[1::4, :]
df_verwitwet_all = df.iloc[2::4, :]
df_geschieden_all = df.iloc[3::4, :]
```

```python
#Creating a list with the columns
alter_list = list(df_all.columns)
```

```python
alter_list = alter_list[6:]
```

```python
for alter in alter_list:
    zero = df_all[['Kanton (-) / Bezirk (>>) / Gemeinde (......)', 'Altersklasse - Total']].copy()
    zero.columns = [['Gemeinde', 'Männer Total']]
    one = df_ledig_all[['Kanton (-) / Bezirk (>>) / Gemeinde (......)', alter]].copy()
    one.columns = [['Gemeinde', alter]]
    two = df_verwitwet_all[['Kanton (-) / Bezirk (>>) / Gemeinde (......)', alter]].copy()
    two.columns = [['Gemeinde', alter]]
    three = df_geschieden_all[['Kanton (-) / Bezirk (>>) / Gemeinde (......)', alter]].copy()
    three.columns = [['Gemeinde', alter]]
    
    zero = zero.merge(one, left_on='Gemeinde', right_on='Gemeinde')
    zero = zero.merge(two, left_on='Gemeinde', right_on='Gemeinde')
    zero = zero.merge(three, left_on='Gemeinde', right_on='Gemeinde')
    zero['Ledige' + alter + 'Männer Total'] = zero.sum(axis=1) - zero['Männer Total']
    zero['Anteil' + alter] = round(zero['Ledige' + alter + 'Männer Total'] / zero['Männer Total'] * 100, 1)
    
    del zero['Männer Total']
    del zero[ alter +'_x']
    del zero[ alter +'_y']
    del zero[ alter ]
    
    df_total_2015 = df_total_2015.merge(zero, left_on='Gemeinde', right_on='Gemeinde')
```
